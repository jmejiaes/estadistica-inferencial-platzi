# Estadistica descriptiva vs inferencial
* Inferencial : parte de la estadística que busca predecir o deducir características o resultados esperados de una población , basados en los datos obtenidos de una muestra de esa población.
* Descriptiva : Tendencia central, variabilidad, distribución
* Inferencial : Abstracción: muestra, intervalos de confianza, validación de hipotesis, tecnicas para evitar sesgos
    * Conclusiones de los parametros de nuestros datos
    * Estudio del grado de fiabilidad de los resultados extraídos del estudio

## Fundamentos de estadistica inferencial

### Uso en data science y machine learning
* Entender la distribución de nuestros datos.
* Crear y validar hipótesis
* hacer experimentos
* elegir los modelos predictivos adecuados según los datos

* Cuantitativo

### Probabilidad 
Mide qué posible es la ocurrencia de un determinado evento

### Probabilidad Condicional
Probabilidad de la ocurrencia de un evento dada la ocurrencia de otro

### Poblaciones Normales e Introducción al teorema del limite central
* Muestreo :  Técnica para la selección de una muestra
* Se obtiene a partir de una población estadística
* La selección tiene que ser aleatoria y se espera que sus propiedades sean extrapolables a la población

#### Tipos de Muestreo
* Aleatorio : Metodos de selección de ciertas unidades sacadas de una población de manera que cada una tiene la misma posibilida de ser elegida
* Sistemático : Método de selección de ciertas unidades al azar, y a continuación se eligen el resto siguiendo intervalos regulares
* Estratificado : Método de selección de ciertas unidades por segmentos exclusivos y homogéneos, para luego usar el muestreo aleatorio simple en cada estrato

#### Teorema del limite central
El teorema del limite central demuestra que de las medias, o de los promedios, de diferentes muestras obtenidas de una misma distribución, va a resultar una distribución normal
Es deir, si de una población se toman varias muestras, y de cada muestra se calcula la media, el conjunto de esas medias va a describir una distribución normal.

#### Muestreo estratificado

```python
econdata['estratificado'] = econdata['delegacion'] + ',' + econdata['tipo']
(econdata['estratificado'].value_counts()/len(econdata)).sort_values(ascending=False)
```

```python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):

  df_estrat = pd.DataFrame(columns = econdata.columns)   ## Se genera un nuevo dataframe vacio con las columnas del dataframe original 

  pos = -1
  for i in range(len(valores_estrat)):    ## Se itera por cada uno de los valores pasados, que serian las proporciones de los estratos
    pos += 1                              
    if pos == len(valores_estrat) - 1:                    # ratio_len: Número de filas que se deben seleccionar para cada estrato.
                                                          # si es la ultima, se asigna la cantidad de observaciones restantes para llenar
                                                          # el df
      ratio_len = len(econdata) - len(df_estrat)
    else:
      ratio_len = int(len(econdata) * prop_estrat[i])     # si no, si asigna la proporcion del estrato, por el tamaño del df

    df_filtrado = econdata[econdata[nombres_columnas_estrat] == valores_estrat[i]]         # df_filtrado: Filtra econdata para obtener solo las filas que coinciden con el valor actual en 
                                                                                           # valores_estrat[i] para las columnas especificadas en nombres_columnas_estrat.
    
    df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)     # Se toman dicha cantidad de valores para la columna de dicho grupo

    df_estrat = pd.concat([df_estrat, df_temp])                                            # Se agregan dichos registros al df que se devolvera
  return df_estrat
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```

#### Tipos de pruebas de hipótesis
* T-student : se usa para estimar una media de población normalmente distribuida a partir de una muestra pequeqlña que sigue  una dist normal, y de la que desconocemos su std
* Coeficiente de pearson: Permte medir la correlación lineal entre dos variables cuantitativas
* Anova :  se usa para compara las varianzas entre las medias (o varianazas) de diferentes grupos

#### Tipos de errores
1. Error tipo I (rechazar h0 cuando es verdadera) (P(Error I) = alpha)
1. Error tipo II (NO rechazar h0 cuando es falsa) (P(Error II) = beta)

#### Prueba con t-student

```python
## Primero debemos calcular el standar error , que es igual a la std / sqrt(n)

se_length = iris['sepal-length'].std() / np.sqrt(len(iris['sepal-length']))
sewidth = iris['sepal-width'].std() / np.sqrt(len(iris['sepal-width']))
```

```python
## Ahora, calculamos el SED, que significa Standar Error of Difference
## su formula es sqrt(se1^2 + se2^2)
sed = np.sqrt(se_length**2 + sewidth**2)
```

```python
## ahora hallamos t, que es igual a la diferencia de las medias / sed
t_stat = (iris['sepal-length'].mean() - iris['sepal-width'].mean()) / sed
t_stat
```

```python
p = ttest_ind(iris['sepal-length'], iris['sepal-width']).pvalue
p
```

### Metodo de bootstrapping
Método de remuestreo con reemplazo
* Muy util en muestras pequqñas oo en distribuciones muy sesgadas
#### ¿Por qué usar bootstrapping?
Porque es un método muy versátil y tenemos acceso al poder de cómputo necesario para hacerlo.

Como recordarás de clases anteriores de este curso, los métodos de t student y de Pearson asumen criterios de la población. Pero esto no siempre es así. No siempre podremos asumir ciertos parámetros y es aquí que bootstrapping se vuelve muy útil porque hace simulaciones de las muestras.

Además, hoy en día es muy accesible la capacidad de cómputo para hacerlo (a menos que sea big data). Bootstrapping es un método que se desarrolló en los 80s, cuando la computación era muy limitada.

Como recomendación, para hacer bootstrapping, asegúrate de tener al menos 25 registros y correr como mínimo 1000 simulaciones. 🚀

```python
import random

data = np.random.normal(loc=34, size=10000)
sns.displot(data, kde=True)
```

```python
data.mean()
indices = random.sample(range(0, len(data)), 20)
pedacito = data[indices]
sns.displot(pedacito,bins=10, kde=True)
```

```python
pedacito.mean()
```

```python
## HAciendo bootstrapping
promedio = []
for i in range(40):
    muestra = random.sample(pedacito.tolist(), 5)
    prom = np.mean(muestra)
    promedio.append(prom)

np.mean(promedio)
```
